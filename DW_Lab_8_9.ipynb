```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
data = pd.read_csv(url)
```

```python
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = df[['total_bill','tip']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.scatter(X['total_bill'], X['tip'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Customer Segmentation Using K-Means Clustering')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df = pd.read_csv('/content/Book1.csv')
```

```python
df
```

```python
X = df[['mid_term_marks','sessional_Marks']]
```

```python
y = df['Final_marks']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
Predictions = model.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,Predictions))
print("R2",r2_score(y_test,Predictions))
```

```python
new_student = pd.DataFrame({'mid_term_marks':[15,25],'sessional_Marks':[20,35]})
```

```python
new_student
```

```python
predict_final_marks = model.predict(new_student)
print('Predicted Final Marks',predict_final_marks)
```
